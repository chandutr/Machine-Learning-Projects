Objective: The main objective of this project is to understand the direct marketing campaigns of a Portuguese banking institution.The scope of the project is to provide knowledge about a big data and to predict if the client will subscribe a term deposit. By applying the analysis over historical data before the regulation, may provide useful insights to solve business issues.The machine learning algorithms are analysed to make the model to get the useful information and to give meaningfull insights.

```python
import pandas as pd
df=pd.read_csv(r'term_deposit.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df = df.rename(columns={'Target': 'deposit'})
```

```python
#to find the specific columns which are having null values
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
df1=df.drop_duplicates
```

```python
for column in df.columns:
    print(column,df[column].nunique())
```

```python
import numpy as np
categorical_columns = [x for x in df.columns if df[x].dtype == np.dtype('object')]
numerical_columns   = [x for x in df.columns if x not in(categorical_columns)]
```

```python
numerical_features=[feature for feature in df.columns if((df[feature].dtypes!='O')&(feature not in ['deposit']))]
print('number of numerical variable:',len(numerical_features))

df[numerical_features].head()
```

```python
df.drop(['default'],axis=1,inplace=True)
```

```python
df.drop(['pdays'],axis=1,inplace=True)
```

```python
#here converting all categorical columns having yes and No into numerical columns
bool_columns = ['housing', 'loan', 'deposit']
existing_bool_cols = [col for col in bool_columns if col in df.columns]

for col in existing_bool_cols:
    df[col] = df[col].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
# performing label encoding on categorical columns using pandas get_dummies function 
#It iterates over the cat_columns list and applies one-hot encoding on each column in B_data4 DataFrame.
import category_encoders as ce

cat_cols = ['job', 'marital', 'education', 'contact', 'month', 'poutcome']
existing_cols = [col for col in cat_cols if col in df.columns]
encoder = ce.OrdinalEncoder(cols=existing_cols)
df = encoder.fit_transform(df)
df.head(2)
```

Split the dataset into input features and target column

```python
x=df.drop(['deposit'], axis=1)
y=df['deposit']
x
```

```python
y
```

Scaling the input feature data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_scaled=pd.DataFrame(x_scaled, columns=x.columns)
x_scaled.head()
```

Divide into Train dataset and test dataset 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42) 
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
y_test.head()
```

Apply Ml into training data

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train, y_train)
```

Model prediction

```python
y_pred=log.predict(x_test)
y_pred[:5]
```

```python
y_test[:5]
```

Model Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test, y_pred)*100
```

```python
print(classification_report(y_test,y_pred))
```
