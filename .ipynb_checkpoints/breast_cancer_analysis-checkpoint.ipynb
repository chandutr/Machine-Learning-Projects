Objective: To create a ML model to predict whether the patient is having breast cancer or not using KNN Algorithm

```python
import pandas as pd
df=pd.read_csv(r'breast-cancer-wisconsin.data.txt', header=None)
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.rename(columns={0:'id', 1:'Clump Thickness', 2:'Uniformity of Cell Size',3:'Uniformity of Cell Shape',4:'Marginal Adhesion',5:'Single Epithelial Cell Size',6:'Bare Nucle',7:'Bland Chromatin',8:'Normal Nucleoli',9:'Mitoses',10:'Class'}, inplace=True)
df
```

```python
df['id'].nunique()
```

```python
df = df.drop('id', axis=1)
```

```python
import numpy as np
df['Bare Nucle'] = df['Bare Nucle'].replace('nan', np.nan)
```

```python
df['Bare Nucle'] = pd.to_numeric(df['Bare Nucle'], errors='coerce')
```

```python
df['Bare Nucle'] = df['Bare Nucle'].fillna(df['Bare Nucle'].mean()).astype(int)
```

Split the dataset into input features and target column.

```python
x=df.drop('Class',axis=1) #input features
y=df['Class'] #output column
x.head()
```

```python
y.head()
```

Scaling the input feature data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_scaled=pd.DataFrame(x_scaled, columns=x.columns)
x_scaled.head()
```

Divide into Train dataset and test dataset 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42) 
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_test.head()
```

```python
y_test.head()
```

Find the optimal K value

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

neighbours=np.arange(1,11)
train_accuracy=[]
test_accuracy=[]

for i in neighbours:
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    train_accuracy.append(knn.score(x_train,y_train))
    test_accuracy.append(knn.score(x_test,y_test))
```

```python
np.arange(1,11)
```

```python
test_accuracy
```

```python
train_accuracy
```

```python
import matplotlib.pyplot as plt
plt.plot(neighbours,train_accuracy, label="train accuracy")
plt.plot(neighbours,test_accuracy, label="test accuracy")
plt.xlabel("neighbors")
plt.ylabel("accuracy")
plt.legend()
```

```python
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

Model Prediction

```python
y_pred=model.predict(x_test)
y_pred[:5]
```

```python
y_test[:5]
```

Model Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm=confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True, fmt='d')
```

```python
accuracy_score(y_test, y_pred)*100
```

```python
print(classification_report(y_test,y_pred))
```

```python
#model building
from sklearn.svm import SVC
mod = SVC(kernel="poly")
mod.fit(x_train,y_train)
```

```python
y_pred=mod.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
#svm with linear
lin_svm=SVC(kernel="linear" , C=1)#c=1
lin_svm.fit(x_train ,y_train)
```

```python
lin_ypred=lin_svm.predict(x_test)
```

```python
lin_acc=accuracy_score(y_test,lin_ypred)
print(lin_acc)
```

```python
#svm with rbf
rbf_svm=SVC(kernel="rbf" , C=100)#c=1
rbf_svm.fit(x_train ,y_train)
```

```python
rbf_ypred=rbf_svm.predict(x_test)
```

```python
rbf_acc=accuracy_score(y_test,rbf_ypred)
print(rbf_acc)
```
