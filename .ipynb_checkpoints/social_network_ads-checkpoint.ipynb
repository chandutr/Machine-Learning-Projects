Objective: To create a ML model to predict purchased price based on Age and Estimated Salary using KNN Algorithm

```python
import pandas as pd
df=pd.read_csv(r'Social_Network_Ads.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
import seaborn as sns 
sns.pairplot(df)
```

Split the dataset into input features and target column.

```python
x=df.drop('Purchased',axis=1) #input features
y=df['Purchased'] #output column
x.head(2)
```

```python
y.head()
```

Scaling the input feature data

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
x_scaled=pd.DataFrame(x_scaled, columns=x.columns)
x_scaled.head()
```

Divide into Train dataset and test dataset 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42) 
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_test.head()
```

```python
y_test.head()
```

Find the optimal K value

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

neighbours=np.arange(1,11)
train_accuracy=[]
test_accuracy=[]

for i in neighbours:
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    train_accuracy.append(knn.score(x_train,y_train))
    test_accuracy.append(knn.score(x_test,y_test))
```

```python
np.arange(1,11)
```

```python
test_accuracy
```

```python
train_accuracy
```

```python
import matplotlib.pyplot as plt
plt.plot(neighbours,train_accuracy, label="train accuracy")
plt.plot(neighbours,test_accuracy, label="test accuracy")
plt.xlabel("neighbors")
plt.ylabel("accuracy")
plt.legend()
```

```python
model=KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
```

Model Prediction

```python
y_pred=model.predict(x_test)
y_pred[:5]
```

```python
y_test[:5]
```

Model Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm=confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot=True, fmt='d')
```

```python
accuracy_score(y_test, y_pred)*100
```

```python
print(classification_report(y_test,y_pred))
```
