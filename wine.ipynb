```python
import pandas as pd
df=pd.read_csv(r'wine.data.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
x=df.drop('Class',axis=1)
y=df['Class']
x.head()
```

Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
x_scaled=pd.DataFrame(x_scaled, columns=x.columns)
x_scaled.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA()
x_pca=pca.fit_transform(x_scaled)
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

sns.barplot(x=range(1,14), y=explained_variance, palette='viridis')
plt.xlabel("Principal Components")
plt.ylabel("Explained Variance")
plt.show()
```

```python
pca_2=PCA(n_components=2)
x_pca=pca_2.fit_transform(x_scaled)
```

```python
x_pca=pd.DataFrame(data=x_pca, columns=['PC1','PC2'])
x_pca
```

Divide into train dataset and test dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=42) 
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

Apply ML algo into training data

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

Model Prediction

```python
y_pred=model.predict(x_test)
y_pred[:5]
```

```python
y_test[:5]
```

Model Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm=confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot=True, fmt='d')
```

```python
accuracy_score(y_test, y_pred)*100
```

```python
print(classification_report(y_test,y_pred))
```
