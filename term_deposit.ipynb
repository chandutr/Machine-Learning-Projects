```python
#here converting all boolean columns having yes and No into numerical columns
bool_columns = ['housing', 'loan', 'deposit']
existing_bool_cols = [col for col in bool_columns if col in df.columns]

for col in existing_bool_cols:
    df[col] = df[col].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
# performing label encoding on categorical columns using pandas get_dummies function 
#It iterates over the cat_columns list and applies one-hot encoding on each column in df DataFrame.
import category_encoders as ce

cat_cols = ['job', 'marital', 'education', 'contact', 'month', 'poutcome']
existing_cols = [col for col in cat_cols if col in df.columns]
encoder = ce.OrdinalEncoder(cols=existing_cols)
df = encoder.fit_transform(df)
df.head(2)
```
